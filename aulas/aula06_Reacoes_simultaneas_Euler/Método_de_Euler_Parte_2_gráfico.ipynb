```
# Importando as bibliotecas
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# constantes das taxas de reação
k1=0.1
#k1=0.001
k2=1.2
k3=0.06
tmax = 10
#dt = 0.1
#dt = 0.01
dt = 0.00000191

# concentração inicial de cada componente química
aold = 20.0
bold = 20.0
cold = 0.0
dold = 0.0
eold = 0.0

tempo = numpy.arange(0,tmax,dt)

t = 0

a = []
b = []
c = []
d = []
e = []
soma = []

while t < tmax:

  RA=-k1 * aold * bold - k3 * aold**2
  RB=-k1 * aold * bold
  RC=k1 * aold * bold - k2 * cold
  RD=k3/2 * aold**2
  RE=2 * k2 * cold

  anew=aold + dt * RA
  bnew=bold + dt * RB
  cnew=cold + dt * RC
  dnew=dold + dt * RD
  enew=eold + dt * RE

  t=t+dt

  aold=anew
  bold=bnew
  cold=cnew
  dold=dnew
  eold=enew
  Sum = aold+bold+cold+dold+eold

  a.append(anew)
  b.append(bnew)
  c.append(cnew)
  d.append(dnew)
  e.append(enew)

# Imprime o passo do tempo e o valor das concentrações no tempo máximo (t = 1)
print("%.8f   %.4f %.4f %.4f %.4f %.4f %.4f" % (dt, anew, bnew, cnew, dnew, enew,Sum))
```

```
print(len(tempo),len(a))
```

```
fig, ax = plt.subplots()
ax.tick_params(axis='both', labelsize=12) #increase font size for ticks
plt.ylabel(r'Concentração (mol/m3)', fontsize=12) #y label
plt.xlabel(r'Tempo (horas)', fontsize=12) #y label
ax.plot(tempo, a, label='Componente A')
ax.plot(tempo, b, label='Componente B')
ax.plot(tempo, c, label='Componente C')
ax.plot(tempo, d, label='Componente D')
ax.plot(tempo, e, label='Componente E')
ax.legend()
#plt.savefig('concentracoes.png', dpi = 300)
plt.show()
```
